```python

def granger_causality_test(df, target, feature, maxlag=12):
    test_results = grangercausalitytests(
        df[[target, feature]], maxlag=maxlag, addconst=True, verbose=False)
    
    results = []

    for lag in range(1, maxlag + 1):
        f_test_pvalue = test_results[lag][0]['ssr_ftest'][1]  # F-test p-value
        chi2_pvalue = test_results[lag][0]['ssr_chi2test'][1]  # Chi-squared p-value
        significant = "Yes" if f_test_pvalue < 0.05 or chi2_pvalue < 0.05 else "No"
        
        results.append({
            'Lag': lag,
            'F-test P-Value': f_test_pvalue,
            'Chi-squared P-Value': chi2_pvalue,
            'Significant': significant
        })

    results_df = pd.DataFrame(results)
    
    return results_df
```

```python

granger_causality_test(df_232, 'pct_blackout', 'wind_speed', maxlag=48)
```
