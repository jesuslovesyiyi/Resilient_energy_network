## Number of variables distribution

```python
not_null_series = bic_df[['pct_blackout_lag', 't2m_lag', 'wind_speed_lag', 'tp_lag']].notnull().sum(axis=1)

x_values = not_null_series.value_counts().index.values
y_values = not_null_series.value_counts().values
plt.figure(figsize=(8,5), dpi=300)
ax = plt.gca()
ax.set_facecolor('xkcd:white')
plt.xticks(np.arange(0, 5, step=1), fontsize=15)
plt.yticks(fontsize=15)
# Bar plot
plt.bar(x_values, y_values, color='grey')
for i in range(len(x_values)):
    plt.text(x_values[i], y_values[i]+1, y_values[i], ha = 'center', fontsize=15)
plt.xlabel('Number of variables', fontsize=20, labelpad=12)
plt.ylabel('Frequency', fontsize=20)
plt.title('Number of variables contributing to ARDL \n Model selection criteria: BIC', fontsize=15)
```
