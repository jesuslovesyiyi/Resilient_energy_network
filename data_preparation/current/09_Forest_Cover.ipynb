```python
import geopandas as gpd

# Load the data
gdf = gpd.read_file('../../data/processed/station_forest_cover.geojson')
```

```python
# plot the statists for columns: 'tree2000', 'gain'

import matplotlib.pyplot as plt

gdf[['tree2000']].plot.hist(bins=50, alpha=0.5, figsize=(10, 6))
plt.title('Histogram of tree2000 and gain')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.show()
```

```python
gdf[['gain']].plot.hist(bins=50, alpha=0.5, figsize=(10, 6))
plt.title('Histogram of  gain')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.show()
```

```python
# show statistics of column 'tree2000' and 'gain'

print(gdf[['tree2000', 'gain']].describe())
```
